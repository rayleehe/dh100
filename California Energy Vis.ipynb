```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
import pandas_alive
import bar_chart_race as bcr
import ffmpeg

df = pd.read_csv("energy.csv")

# Replace 'California' string with the state you wish to look into.
df_california = df[df['State'] == 'California']
```

```python
# Replace these column labels with the data you want to look at, this may include different consumption sources,
# production numbers, usage figures, and more.
df_california_specific = df_california[['Year', 'Consumption.Commercial.Coal', 
                                       'Consumption.Commercial.Petroleum',
                                       'Consumption.Commercial.Natural Gas',
                                       'Consumption.Commercial.Solar',
                                       'Consumption.Commercial.Wind',
                                       'Consumption.Commercial.Hydropower']]
#df_california_specific['Year'] = pd.to_datetime(df_california_specific['Year'])
#display(df_california_specific)

# Use the "Year" column as the index
df_california_specific = df_california_specific.set_index("Year")
df_california_specific = df_california_specific.rename(columns={'Consumption.Commercial.Coal':'Coal',
                                                               'Consumption.Commercial.Petroleum':'Petroleum',
                                                               'Consumption.Commercial.Natural Gas':'Natural Gas',
                                                               'Consumption.Commercial.Solar':'Solar',
                                                               'Consumption.Commercial.Wind':'Wind',
                                                               'Consumption.Commercial.Hydropower':'Hydropower'})
df_california_specific.head(5)
```
